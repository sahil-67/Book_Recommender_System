## Popularity Based Recommender System

```python
# We will consider top 250 books based on ratings. In case of a tie, minimum 250 ratings on a book will be considered
```

```python
ratings
```

```python
books
```

##### Merging 'books' & 'ratings' table on top of ISBN

```python
ratings_with_name = ratings.merge(books , on = 'ISBN' )
```

```python
ratings_with_name.head()
```

```python
num_rating_df =ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
```

```python
num_rating_df
```

```python
num_rating_df.rename(columns={'Book-Rating':'num_ratings'} , inplace=True)
num_rating_df
```

```python
avg_rating_df =ratings_with_name.groupby('Book-Title').mean()['Book-Rating'].reset_index()
```

```python
avg_rating_df.rename(columns={'Book-Rating':'avg_rating'},inplace=True)
avg_rating_df
```

```python
popularity_df = num_rating_df.merge(avg_rating_df , on='Book-Title')
popularity_df
```

```python
popularity_df = popularity_df[popularity_df['num_ratings']>=250].sort_values('avg_rating' , ascending=False).head(50)
```

```python
popularity_df.merge(books , on ='Book-Title').drop_duplicates('Book-Title')
```

## Collaborative Based Filtering

```python
# We will create a table in which columns will have user attribute while rows will have book attribute.

# The intersection of a column and row will indicate the rating done on a book by a user

# For this model, we will consider only those users who have done a minimum of 200 ratings(experienced user)

# We will consider only those books on which minimum 50 ratings have been done(not considering unpopular books)
```

```python
ratings_with_name.groupby('Book-Title').sum().reset_index()
```

```python
x = ratings_with_name.groupby('User-ID').count()['Book-Rating'] > 200
u200 = x[x].index
```

```python
filtered_rating = ratings_with_name[ratings_with_name['User-ID'].isin(u200)]
```

```python
filtered_rating.columns
```

```python
filtered_rating.head()
```

```python
y = filtered_rating.groupby('Book-Title').count()['Book-Rating'] >=50
famous_books = y[y].index
```

```python
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
final_ratings.duplicated().sum()
```

```python
pt = final_ratings.pivot_table(index = 'Book-Title' , columns='User-ID', values = 'Book-Rating')
```

```python
pt
```

```python
pt.fillna(0, inplace=True)
```

```python
pt
```

```python
# Now we will assume each book in a 810 dimensional field and then find Eucledian distance bewteen all books
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score = cosine_similarity(pt)
```

```python
np.where(pt.index=='You Belong To Me')[0][0]
```

```python
sorted(list(enumerate(similarity_score[9])), reverse=True , key=lambda x:x[1])
```

```python
def recommender(book_name):
    index = np.where(pt.index==book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_score[9])), reverse=True , key=lambda x:x[1])[1:6]


    for i in similar_items:
        print(pt.index[i[0]])
```
